```python
# Selecting only the columns containing survey responses
survey_data = df_filtered.loc[:, 'Q1.1':'Q1.29']
```

```python
# Initialize factor analyzer
fa = FactorAnalyzer()

# Fit factor analysis model to calculate eigenvalues
fa.fit(survey_data)

# Get eigenvalues
eigenvalues = fa.get_eigenvalues()
```

```python
# Number of factors to extract
num_factors = 5  # You can adjust this number based on your analysis
```

```python
# Create factor analyzer object
fa = FactorAnalyzer(n_factors=num_factors, rotation=None)  # Setting rotation to None for simplicity
```

```python
# Fit the model
fa.fit(survey_data)
```

```python
# Get factor loadings
factor_loadings = fa.loadings_
```

```python
# Display factor loadings
print("Factor Loadings:")
print(factor_loadings)
```

```python
print("Communalities:")
print(fa.get_communalities())
```

```python
factor_scores = survey_data.dot(factor_loadings)
factor_scores
```

```python
dfid = df['RESPID']
dfid_df = pd.DataFrame(dfid, columns=['RESPID'])
```

```python
dfid_df
```

```python
combined_df = pd.concat([dfid_df, factor_scores], axis=1)
combined_df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Extract factor scores from combined_df

# Scale the factor scores
scaler = StandardScaler()
scaled_factor_scores = scaler.fit_transform(factor_scores)
```

```python
# Initialize list to store within-cluster sum of squares
wcss = []

# Define range of clusters to try
k_range = range(1, 20)  # Try k values from 1 to 10

# Calculate WCSS for each k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_factor_scores)
    wcss.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.metrics import silhouette_score
```

```python
# Define range of clusters to try
k_range = range(2, 20)  # Try k values from 2 to 10

# Initialize list to store silhouette scores
silhouette_scores = []

# Calculate silhouette score for each k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_factor_scores)
    silhouette_avg = silhouette_score(scaled_factor_scores, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Scores for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
#5 is optimal k
```

```python

# Define the number of clusters
num_clusters = 5  # Adjust as needed

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(factor_scores)

# Assign cluster labels to original data
combined_df['cluster'] = cluster_labels
```

```python

# Check the distribution of clusters
print(combined_df['cluster'].value_counts())
```

```python

# Explore the characteristics of each cluster
cluster_means = combined_df.groupby('cluster').mean()
print(cluster_means)
```
